```python
import pandas as pd


# Load Excel
df = pd.read_excel("Online Retail.xlsx", engine="openpyxl", sheet_name=0)

# quick checks
print("Shape:", df.shape)
display(df.head())
df.info()
```

```python
# 1) columns
print(df.columns.tolist())

# 2) first rows
df.head()

# 3) basic info
df.info()

# 4) summary
df.describe(include='all')
```

```python
# Shape of dataset
print("Rows, Columns:", df.shape)

# Column names
print(df.columns)

# Basic structure
df.info()

# First 5 rows
df.head()
```

```python
df.isnull().sum()
```

```python
# Drop rows without CustomerID
df = df.dropna(subset=['CustomerID'])

print("Shape after removing null CustomerID:", df.shape)
```

```python
df.isnull().sum()
```

```python
# Keep only positive quantities
df = df[df['Quantity'] > 0]

print("Shape after removing negative Quantity:", df.shape)
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df[['Quantity', 'UnitPrice', 'TotalPrice']].head()
```

```python
# Group by CustomerID and calculate useful metrics
customer_df = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',      # number of unique invoices (purchases)
    'Quantity': 'sum',           # total quantity purchased
    'TotalPrice': 'sum'          # total spending
}).reset_index()

# Rename columns for clarity
customer_df.rename(columns={
    'InvoiceNo': 'NumPurchases',
    'Quantity': 'TotalQuantity',
    'TotalPrice': 'TotalSpend'
}, inplace=True)

# Preview
print(customer_df.shape)
customer_df.head()
```

```python
# Info & summary
print(customer_df.info())
print(customer_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(18,5))

sns.histplot(customer_df['NumPurchases'], bins=30, kde=True, ax=axes[0], color='blue')
axes[0].set_title("Number of Purchases per Customer")

sns.histplot(customer_df['TotalQuantity'], bins=30, kde=True, ax=axes[1], color='green')
axes[1].set_title("Total Quantity Purchased")

sns.histplot(customer_df['TotalSpend'], bins=30, kde=True, ax=axes[2], color='red')
axes[2].set_title("Total Spending per Customer")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(customer_df[['NumPurchases','TotalQuantity','TotalSpend']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Customer Metrics")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,5))

sns.boxplot(y=customer_df['NumPurchases'], ax=axes[0], color='blue')
axes[0].set_title("NumPurchases per Customer")

sns.boxplot(y=customer_df['TotalQuantity'], ax=axes[1], color='green')
axes[1].set_title("TotalQuantity per Customer")

sns.boxplot(y=customer_df['TotalSpend'], ax=axes[2], color='red')
axes[2].set_title("TotalSpend per Customer")

plt.show()
```

```python
X = customer_df[['NumPurchases', 'TotalQuantity', 'TotalSpend']]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# check scaled values
import pandas as pd
pd.DataFrame(X_scaled, columns=X.columns).head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

# Try k=1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot elbow
plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method to find optimal k")
plt.show()
```

```python
# Suppose elbow shows optimal k=4
kmeans = KMeans(n_clusters=4, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(X_scaled)

# Preview cluster assignment
customer_df.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=customer_df, x='TotalSpend', y='TotalQuantity', hue='Cluster', palette='Set2', s=100)
plt.title("Customer Segments")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(customer_df['NumPurchases'], customer_df['TotalQuantity'], customer_df['TotalSpend'],
           c=customer_df['Cluster'], cmap='Set2', s=50)
ax.set_xlabel('NumPurchases')
ax.set_ylabel('TotalQuantity')
ax.set_zlabel('TotalSpend')
plt.title("3D View of Customer Segments")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(customer_df['NumPurchases'], customer_df['TotalQuantity'], customer_df['TotalSpend'],
           c=customer_df['Cluster'], cmap='Set2', s=50)
ax.set_xlabel('NumPurchases')
ax.set_ylabel('TotalQuantity')
ax.set_zlabel('TotalSpend')
plt.title("3D View of Customer Segments")
plt.show()
```

```python
# Aggregate stats per cluster
cluster_summary = customer_df.groupby('Cluster').agg({
    'NumPurchases': ['mean', 'min', 'max'],
    'TotalQuantity': ['mean', 'min', 'max'],
    'TotalSpend': ['mean', 'min', 'max'],
    'CustomerID': 'count'  # number of customers in cluster
}).round(2)

print(cluster_summary)
```

```python
import matplotlib.pyplot as plt
cluster_counts = customer_df['Cluster'].value_counts().sort_index()
plt.figure(figsize=(6,4))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette="Set2")
plt.xlabel("Cluster")
plt.ylabel("Number of Customers")
plt.title("Number of Customers per Cluster")
plt.show()
```

```python
sns.scatterplot(data=customer_df, x='TotalSpend', y='TotalQuantity', hue='Cluster', palette='Set2', s=100)
plt.title("Customer Segmentation: Spend vs Quantity")
plt.show()
```
